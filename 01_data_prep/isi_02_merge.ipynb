### 4) Spalten löschen

```python
auftraege_auftragsdetails_aufenthalt_stammdaten = auftraege_auftragsdetails_aufenthalt_stammdaten.drop(['AKTIV', 'AUSMUSTERUNG_DAT', 'MONAT', 'JAHR',"BAUART_y","ISAS","UPDATE_DATE"], axis=1)

#auftraege_auftragsdetails_aufenthalt_stammdaten.UPDATE_DATE = auftraege_auftragsdetails_aufenthalt_stammdaten.UPDATE_DATE.astype(int)
auftraege_auftragsdetails_aufenthalt_stammdaten.INBETRIEB_AB = auftraege_auftragsdetails_aufenthalt_stammdaten.INBETRIEB_AB.astype(int)
auftraege_auftragsdetails_aufenthalt_stammdaten.BAUJAHR = auftraege_auftragsdetails_aufenthalt_stammdaten.BAUJAHR.astype(int)

auftraege_auftragsdetails_aufenthalt_stammdaten = auftraege_auftragsdetails_aufenthalt_stammdaten.rename(columns={'BAUART_x': 'BAUART'})
```

### 5) Baujahr-Datum in Excel-Format umwandeln

```python
auftraege_auftragsdetails_aufenthalt_stammdaten = auftraege_auftragsdetails_aufenthalt_stammdaten.drop(auftraege_auftragsdetails_aufenthalt_stammdaten[(auftraege_auftragsdetails_aufenthalt_stammdaten["BAUJAHR"] < 1800)].index)

auftraege_auftragsdetails_aufenthalt_stammdaten["BAUJAHR"] = auftraege_auftragsdetails_aufenthalt_stammdaten["BAUJAHR"].apply(lambda x: dt.datetime(x,month=1,day=2))

def datetime_as_xldate(pydatetime):
    temp = dt.datetime(1899, 12, 31)
    delta = pydatetime - temp
    return float(delta.days) + (float(delta.seconds) / 86400)

auftraege_auftragsdetails_aufenthalt_stammdaten["BAUJAHR"] = auftraege_auftragsdetails_aufenthalt_stammdaten["BAUJAHR"].apply(datetime_as_xldate)

auftraege_auftragsdetails_aufenthalt_stammdaten.BAUJAHR = auftraege_auftragsdetails_aufenthalt_stammdaten.BAUJAHR.astype(int)
```

### 6) NA-Werte, Plätze und Türen mit Eintrag 0 und unrealistischen Einträgen löschen

```python
auftraege_auftragsdetails_aufenthalt_stammdaten = auftraege_auftragsdetails_aufenthalt_stammdaten.drop(auftraege_auftragsdetails_aufenthalt_stammdaten[(auftraege_auftragsdetails_aufenthalt_stammdaten["PLAETZE"] == 0)].index)
auftraege_auftragsdetails_aufenthalt_stammdaten = auftraege_auftragsdetails_aufenthalt_stammdaten.drop(auftraege_auftragsdetails_aufenthalt_stammdaten[(auftraege_auftragsdetails_aufenthalt_stammdaten["PLAETZE"] == 1)].index)
auftraege_auftragsdetails_aufenthalt_stammdaten = auftraege_auftragsdetails_aufenthalt_stammdaten.drop(auftraege_auftragsdetails_aufenthalt_stammdaten[(auftraege_auftragsdetails_aufenthalt_stammdaten["PLAETZE"] == 2)].index)
auftraege_auftragsdetails_aufenthalt_stammdaten = auftraege_auftragsdetails_aufenthalt_stammdaten.drop(auftraege_auftragsdetails_aufenthalt_stammdaten[(auftraege_auftragsdetails_aufenthalt_stammdaten["PLAETZE"] == 3)].index)
auftraege_auftragsdetails_aufenthalt_stammdaten = auftraege_auftragsdetails_aufenthalt_stammdaten.drop(auftraege_auftragsdetails_aufenthalt_stammdaten[(auftraege_auftragsdetails_aufenthalt_stammdaten["PLAETZE"] == 4)].index)
auftraege_auftragsdetails_aufenthalt_stammdaten = auftraege_auftragsdetails_aufenthalt_stammdaten.drop(auftraege_auftragsdetails_aufenthalt_stammdaten[(auftraege_auftragsdetails_aufenthalt_stammdaten["PLAETZE"] == 5)].index)

auftraege_auftragsdetails_aufenthalt_stammdaten = auftraege_auftragsdetails_aufenthalt_stammdaten.dropna(subset=['FZ_KATEGORIE','MOTORTYP','ABGASFILTER','EURONORM'])

auftraege_auftragsdetails_aufenthalt_stammdaten = auftraege_auftragsdetails_aufenthalt_stammdaten.drop(auftraege_auftragsdetails_aufenthalt_stammdaten[(auftraege_auftragsdetails_aufenthalt_stammdaten["ANZAHL_TUEREN"] == 0)].index)
```
